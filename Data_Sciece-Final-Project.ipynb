# <a id="5">  5 ) . What is the distribution for the trip duration in minutes? </a>

```python
## Review the statistic for trip duration in minutes

df['duration_min'].describe()
```

```python
# Make visualization to easy get insight


if 'duration_min' not in df.columns:
    df['duration_min'] = df['duration_sec'] / 60 

plt.figure(figsize=(12, 6))
sns.histplot(df['duration_min'], kde=True, color='teal', bins=50, stat='density')

plt.title("Distribution of Trip Duration in Minutes", fontsize=18)
plt.xlabel("Duration (minutes)", fontsize=14)
plt.ylabel("Density", fontsize=14)

plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

> ### <mark>The graph shows that most trips are very short (less than 50 minutes), with a steep decline in density as duration increases. A few outliers exist beyond 200 minutes, but the data is heavily skewed toward shorter durations. </mark>

```python
plt.figure(figsize=(12, 6))

sns.violinplot(x='user_type', y='duration_min', hue='member_gender', data=df, split=True, palette='Set1')

plt.title("Distribution of Trip Duration by User Type and Gender", fontsize=18, fontweight='bold')

plt.show()
```

```python
corr = df[['age', 'duration_min']].corr()


plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

plt.title("Correlation Matrix between Age and Trip Duration", fontsize=18, fontweight='bold')

plt.show()
```

```python
g = sns.FacetGrid(df, col='user_type', row='member_gender', margin_titles=True)
g.map(sns.scatterplot, 'age', 'duration_min', alpha=0.7)

g.fig.suptitle("Relationship between Age and Trip Duration by User Type and Gender", fontsize=18, fontweight='bold')

plt.show()
```
